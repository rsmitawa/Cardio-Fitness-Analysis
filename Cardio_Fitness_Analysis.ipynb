## Visualizations and tables

##### Univariate Analysis

```python
# Unique Product count in the dataset
tb=cardio_df['Product'].value_counts()
print(tb)
ax = tb.plot(kind='bar')
ax.set_title('Bar chart of the frequency of treadmills model')
plt.show()
```

```python
# Histogram of variable Age
ax = sns.histplot(x = 'Age',  data=cardio_df)
ax.set_title('Histogram to check distribution of Age')
plt.show()
```

The age of customers follows a Right-tailed Normal distriution curve with the mean score of 28.8 and standard deviation of 6.94.

```python
# Table and bar chart of variable Gender
tb=cardio_df['Gender'].value_counts()
print(tb)
ax = tb.plot(kind='bar')
ax.set_title('Bar chart of Gender of the customer')
plt.show()
```

Sample consists of 104 males and 76 females.

```python
# Bar chart of variable Education
ax = cardio_df['Education'].value_counts(sort=False).plot(kind='bar')
ax.set_title('Bar chart of Education of the customer')
ax.set(xlabel='Education (years)', ylabel='Frequency')
plt.show()
```

###### Insights

Most of the customers has 14 or 16 years of education experience. We can conclude that most of the customers are graduates and undergraduates. 

```python
# Plotting a pie chart for marital status

# Frequency table of marital status
marital_table = cardio_df['MaritalStatus'].value_counts()

# Function to calculate percentage of population for labeling
def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Plotting pie chart using matplotlib
plt.pie(marital_table, 
        labels = marital_table.index,
        autopct=lambda pct: func(pct, marital_table), 
        shadow=True)
plt.title("Pie chart for marital status")
plt.show()
```

###### Insight

The pie chart shows the population demographic of customers with their marital status.

```python
# Histogram of Avg. no of times the customer wants to use the treadmill every week
ax = sns.histplot(x = 'Usage', data=cardio_df)
ax.set_title('Histogram to check distribution of Usage')
plt.show()
```

###### Insights and Recommendations

1. Mostly customers wants to use treadmill 3-4 times in a week. A daywise machine usage can be calculated to optimise the usage of machines for these customers by collecting their preference data.
2. Very few people wants to use treadmill everyday.

```python
# Frequency bar chart of self-rated fitness score of the customer
ax = cardio_df['Fitness'].value_counts(sort=False).plot(kind='bar')
ax.set_title('Bar chart of Fitness of the customer')
ax.set(xlabel='Fitness ( 1 - very unfit, 5 - very fit)')
plt.show()
```

###### Insights

Most of the people consider themselves average fit and very few consider very unfit.

```python
# Histogram of income of the customer
ax=sns.displot(cardio_df['Income'], kde=True)
plt.show()
```

###### Insights

The customers have average income of around 50k ranging from around 30k to 105k per year. The distribution is right tailed and many people comes in high range salary.

```python
# Histogram of distance in miles that a customer expects to run
ax=sns.displot(cardio_df['Miles'], kde=True)
ax.set_axis_labels('Expected run in Miles')
plt.show()
```

###### Insights

Most of the sample population of customers expects to run 103 miles on average, and some of them have a target of of more than 200 miles, twice of average customers. 

##### Bivariate Analysis

```python
# Bar chart of Usage for each product gender wise
ax = sns.barplot(x = "Product", y = 'Usage', hue = 'Gender', data=cardio_df)
ax.set_title('Avg. number times the customer wants to use the treadmill every week on gender level')
plt.show()
```

###### Insights and Recommendations

1. Males slightly prefer more TM195 treadmill and females prefers TM798 compared to males.
2. Treadmill TM798 has high preference compared to other machines overall.
3. We can checkout the distinct features of TM798 for more information to attract more customers.

```python
# Bar chart of Usage for each product on Marital Status
ax = sns.barplot(x = "Product", y = 'Usage', hue = 'MaritalStatus', data=cardio_df)
ax.set_title('Avg. number times the customer wants to use the treadmill every week on Marital Status basis')
plt.show()
```

###### Insights
No special prefernces can be seen for single or partnered people for treadmills.

```python
# Boxplot of Fitness for all age groups people
ax = sns.boxplot(x = "Fitness", y = 'Age', data=cardio_df)
ax.set_title('Fitness for all age groups people in boxplot')
plt.show()
```

###### Insights

Young people from age 25-30 years old rated themselves highly unfit compared to higher age group. Their defination of fitness might be different from higher age group people.

```python
# Bar chart of Fitness and incomes of customers
ax = sns.barplot(x = "Fitness", y = 'Income',data=cardio_df)
ax.set_title('Self rated fitness score of the customer with their income')
plt.show()
```

###### Insights

People with high income have rated themsevles highly fit.

```python
# Bar chart of Fitness with running miles per gender
ax = sns.barplot(x = "Fitness", y = 'Miles',hue = 'Gender',data=cardio_df)
ax.set_title('Self rated fitness score with running miles per gender')
plt.show()
```

###### Insights and recommendations

1. Higher the fitness on self-fitness scale, higher the number of miles a person wants to run and use treadmill.
2. Self-rated highly fit women expects to run more than 200 miles, whereas males don't want to run more than 170 miles.
3. Females who have rated themselves highly unfit, don't want to run on treadmill. We can further investigate into this more with more data about their fitness goal.

```python
# Scatterplot of income and running miles of customer coloured with marital status
ax=sns.scatterplot(x='Income', y='Miles', hue='MaritalStatus', data=cardio_df)
ax.set_title("Income of custore and running miles with Marital status")
plt.show()
```

###### Insights
